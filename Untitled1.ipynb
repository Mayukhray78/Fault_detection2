```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('Output (S)',axis=1),df['Output (S)'],random_state=42,test_size=0.2)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Define a learning rate scheduler
def lr_scheduler(epoch, lr):
    # Warm-up for the first 5 epochs, then use cosine decay
    if epoch < 5:
        return lr + 0.0005
    else:
        return 0.0001 * 0.5 * (1 + np.cos(np.pi * epoch / 50))
```

```python
# Updated model architecture
model = Sequential([
    Dense(256, input_dim=x_train.shape[1]),   # First dense layer
    BatchNormalization(),                    # Batch normalization for stability
    LeakyReLU(alpha=0.1),                    # LeakyReLU to avoid dead neurons
    Dropout(0.5),                            # Dropout for regularization

    Dense(128),                              # Second dense layer
    BatchNormalization(),                    # Batch normalization
    LeakyReLU(alpha=0.1),                    # LeakyReLU activation
    Dropout(0.5),                            # Dropout for regularization

    Dense(64),                               # Third dense layer
    BatchNormalization(),                    # Batch normalization
    LeakyReLU(alpha=0.1),                    # LeakyReLU activation
    Dropout(0.5),                            # Dropout for regularization

    Dense(1, activation='sigmoid')           # Output layer for binary classification
])
```

```python
from keras.optimizers import Adam

# Compile the model using RAdam optimizer (2019) with a learning rate scheduler
optimizer = Adam(learning_rate=0.001)  # RAdam optimizer
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Adding callbacks for EarlyStopping and LearningRateScheduler
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
lr_callback = LearningRateScheduler(lr_scheduler)
```

```python
# Train the model
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=100, batch_size=32, 
                    callbacks=[early_stopping, lr_callback])
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(10,3))
ax1.plot(history.history['loss'],label='training')
ax1.plot(history.history['val_loss'],label='validation')
ax1.set_title('loss')
ax1.set_xlabel('epoch')
ax1.set_ylabel('loss')

ax2.plot(history.history['accuracy'])
ax2.plot(history.history['val_accuracy'])
ax2.set_title('accuracy')
ax2.set_xlabel('epoch')
ax2.set_ylabel('accuracy')

plt.subplots_adjust(wspace=0.3)
fig.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=2)
plt.show()
```

```python
# Calculate predictions
y_pred = (model.predict(x_test) > 0.5).astype("int32")

# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
